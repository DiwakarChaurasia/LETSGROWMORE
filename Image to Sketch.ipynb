```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.imshow(image)
plt.title("image",fontstyle='italic', fontsize=10,color="gray")
plt.axis(False)
plt.show()
```

### Loading RGB image

```python
RGB_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_image)
plt.title("RGB_Image",fontstyle='italic', fontsize=10,color="gray")
plt.axis(False)
plt.show()
```

### Loading Gray Image

```python
GrayScale_image = cv2.cvtColor(RGB_image, cv2.COLOR_RGB2GRAY)
plt.imshow(GrayScale_image)
plt.title("Gray_Scale_Image",fontstyle='italic', fontsize=10,color="gray")
plt.axis(False)
plt.show()
```

### Blending the grayscale and blurred grayscale images

```python
# Displays the grayscale and blurred grayscale image
RGBSketch_image = cv2.divide(GreyScale_image, InvertedBlur_image, scale=256.0)
plt.imshow(RGBSketch_image)
plt.title("RGB_Sketch_Image",fontstyle='italic', fontsize=10,color="gray")
plt.axis(False)
plt.show()
```

### Displaying the Pencil sketch image

```python
PencilSketch_image = cv2.cvtColor(RGBSketch_image, cv2.COLOR_RGB2BGR)
plt.figure(figsize = (8,8))
plt.title("Pencil_Sketch_Image",fontstyle='italic', fontsize=20,color="gray")
plt.imshow(PencilSketch_image)
plt.axis(False)
plt.show()
```
